```python
mv_zip_pre = pd.DataFrame(columns = ['ZIPCODE_mv','PRECINCT_mv'])
#take mode of zipcode from 'INCIDENT_DISPATCH_AREA' to replace all zipcode NaN
#do we take from Data or Data_csv that has deleted Nans? 
mv_zip_pre['ZIPCODE_mv'] = data_csv.groupby('INCIDENT_DISPATCH_AREA')['ZIPCODE'].apply(lambda x: x.value_counts().index[0])
mv_zip_pre['PRECINCT_mv'] = data_csv.groupby('INCIDENT_DISPATCH_AREA')['POLICEPRECINCT'].apply(lambda x: x.value_counts().index[0])
mv_zip_pre
```

```python
data_clean = data_csv.merge(mv_zip_pre,on = 'INCIDENT_DISPATCH_AREA')
data_clean
```

```python
data_clean.loc[data_clean['ZIPCODE'].isnull(),['ZIPCODE']] = data_clean["ZIPCODE_mv"]
data_clean.loc[data_clean['POLICEPRECINCT'].isnull(),['POLICEPRECINCT']] = data_clean["PRECINCT_mv"]
```

```python
mv_district = pd.DataFrame(columns = ['CITYCOUNCILDISTRICT_mv','COMMUNITYDISTRICT_mv',
                                      "COMMUNITYSCHOOLDISTRICT_mv","CONGRESSIONALDISTRICT_mv"])
#take mode of zipcode from 'INCIDENT_DISPATCH_AREA' to replace all zipcode NaN
#do we take from Data or district that has deleted Nans? 
mv_district['CITYCOUNCILDISTRICT_mv'] = data_csv.groupby('INCIDENT_DISPATCH_AREA')['CITYCOUNCILDISTRICT'].apply(lambda x: x.value_counts().index[0])
mv_district['COMMUNITYDISTRICT_mv'] = data_csv.groupby('INCIDENT_DISPATCH_AREA')['COMMUNITYDISTRICT'].apply(lambda x: x.value_counts().index[0])
mv_district['COMMUNITYSCHOOLDISTRICT_mv'] = data_csv.groupby('INCIDENT_DISPATCH_AREA')["COMMUNITYSCHOOLDISTRICT"].apply(lambda x: x.value_counts().index[0])
mv_district['CONGRESSIONALDISTRICT_mv'] = data_csv.groupby('INCIDENT_DISPATCH_AREA')['CONGRESSIONALDISTRICT'].apply(lambda x: x.value_counts().index[0])
```
